```python
output_path = 'C:/Users/User/Desktop/CS450/Data Set NS 20/Augmented_train_data_mixed3X.csv'
```

```python
import pandas as pd

# เลือกคอลัมน์ที่เราจะใช้

cols_keep = ["review_body", "labels", "is_synthesis"]

df1_subset = df1[cols_keep].copy()
df2_subset = df2[cols_keep].copy()


# STEP 1: เอา real data (is_synthesis == 0) จาก df1 เท่านั้น


real_rows = df1_subset[df1_subset["is_synthesis"] == 0]

# STEP 2: เอา synthetic data (is_synthesis == 1)
#         จากทั้ง df1 และ df2

syn_df1 = df1_subset[df1_subset["is_synthesis"] == 1]
syn_df2 = df2_subset[df2_subset["is_synthesis"] == 1]

synthetic_rows = pd.concat([syn_df1, syn_df2], ignore_index=True)


# STEP 3: รวม real + synthetic (จากทุกแหล่ง)

combined_df = pd.concat(
    [real_rows, synthetic_rows],
    ignore_index=True
)


# STEP 4: export CSV

combined_df.to_csv(output_path, index=False, encoding="utf-8-sig")

print("Final shape:", combined_df.shape)
print("Real rows   :", (combined_df["is_synthesis"] == 0).sum())
print("Synthetic rows:", (combined_df["is_synthesis"] == 1).sum())
```
